```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# SPECTROGRAM WITH WHITE BORDERS

data1= np.load('/content/drive/Shareddrives/RAVEDESS/MODEL/spec train white.npy')
data2=np.load('/content/drive/Shareddrives/RAVEDESS/MODEL/spec test white.npy')
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)


# Changing labels to integer value to fit to label
df1[25088] = df1[25088].replace('8','0')  #(label number 0 represents surprise emotion(08))
df2[25088] = df2[25088].replace('8','0')  #(label number 0 represents surprise emotion(08))

# EXTRACTING DEPENDENT AND INDEPENDENT VARIABLE
X_train = df1.iloc[:, :-1].values
Y_train = df1.iloc[:, 25088].values
X_test = df2.iloc[:, :-1].values
Y_test = df2.iloc[:, 25088].values
```

```python
# USING RANDOM FOREST TO PREDICT EMOTION
#--------------------------------------------------------------------->>>

clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,Y_train)
rf_pred=clf.predict(X_test)
rf_score=(accuracy_score(Y_test,rf_pred))*100
print('Accuracy Score:',rf_score)
```

```python
# USING DECISION TREE TO PREDICT EMOTION
# ---------------------------------------------------------------->>>

classifier = DecisionTreeClassifier()
classifier = classifier.fit(X_train,Y_train)
genre = classifier.predict(X_test)
dec_res=(accuracy_score(Y_test,genre))*100
print('Accuracy Score:',dec_res)
```

```python
# SVM
# Making the SVM Classifer
Classifier = SVC(kernel="linear")

# Training the model on the training data and labels
Classifier.fit(X_train, Y_train)

# Using the model to predict the labels of the test data
Y_pred = Classifier.predict(X_test)

# Evaluating the accuracy of the model using the sklearn functions
accuracy = (accuracy_score(Y_test,Y_pred))*100

# Printing the results
print("Accuracy for SVM is:",accuracy)
```

```python
# Using XGBoost
# ------------------------------------------------------------->>>

# Applying XGBoost
xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(X_train,Y_train)
preds = xgb_classifier.predict(X_test)
xgb_score=(accuracy_score(Y_test,preds))*100
print('Accuracy Score : ',xgb_score)
```

```python
# SPECTROGRAM WITHOUT WHITE BORDERS

data3= np.load('/content/drive/Shareddrives/RAVEDESS/MODEL/spec train.npy')
data4=np.load('/content/drive/Shareddrives/RAVEDESS/MODEL/spec test.npy')
df3=pd.DataFrame(data3)
df4=pd.DataFrame(data4)

# Changing labels to integer value to fit to label
df3[18432] = df3[18432].replace('8','0')  #(label number 0 represents surprise emotion(08))
df4[18432] = df4[18432].replace('8','0')  #(label number 0 represents surprise emotion(08))

# EXTRACTING DEPENDENT AND INDEPENDENT VARIABLE
x_train = df3.iloc[:, :-1].values
y_train = df3.iloc[:, 18432].values
x_test = df4.iloc[:, :-1].values
y_test = df4.iloc[:, 18432].values
```

```python
# USING RANDOM FOREST TO PREDICT GENRE
#--------------------------------------------------------------------->>>

comps = [25, 50, 100, 150]

for n_comps in comps:

    pca = PCA(n_components = n_comps)

    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)

    clf=RandomForestClassifier(n_estimators=100)
    clf.fit(x_train_pca,y_train)
    rf_pred=clf.predict(x_test_pca)
    rf_score1=(accuracy_score(y_test,rf_pred))*100
    print('Accuracy Score:',rf_score1)
```

```python
# USING DECISION TREE TO PREDICT GENRE
# ---------------------------------------------------------------->>>

comps = [150, 250, 400, 500, 750, 1000]

for n_comps in comps:

    pca = PCA(n_components = n_comps)

    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)

    classifier = DecisionTreeClassifier()
    classifier = classifier.fit(x_train_pca,y_train)
    genre = classifier.predict(x_test_pca)
    dec_res1=(accuracy_score(y_test,genre))*100
    print('Accuracy Score:',dec_res1)
```

```python
comps = [150, 250, 400, 500, 750, 1000]

for n_comps in comps:

    pca = PCA(n_components = n_comps)

    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)

    Classifier = SVC(kernel="rbf")
    Classifier.fit(x_train_pca, y_train)
    y_pred = Classifier.predict(x_test_pca)
    accuracy1 = (accuracy_score(y_test,y_pred))*100
    print("Accuracy for SVM is:",accuracy1, "with n components as", n_comps)
```

```python
# Using XGBoost
# ------------------------------------------------------------->>>

# Applying XGBoost

comps = [25, 50, 100, 150, 250, 400]

for n_comps in comps:

    pca = PCA(n_components = n_comps)

    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)

    xgb_classifier = xgb.XGBClassifier()
    xgb_classifier.fit(x_train_pca,y_train)
    preds = xgb_classifier.predict(x_test_pca)
    xgb_score1=(accuracy_score(y_test,preds))*100
    print('Accuracy Score : ',xgb_score1)
```

```python
# LIBROSA EXTRACTED FEATURES

df5=pd.read_csv('/content/drive/Shareddrives/RAVEDESS/MODEL/librosa_train.csv')
df6=pd.read_csv('/content/drive/Shareddrives/RAVEDESS/MODEL/librosa_test.csv')

# Changing labels to integer value to fit to label
df5['emotion'] = df5['emotion'].replace('8','0')  #(label number 0 represents surprise emotion(08))
df6['emotion'] = df6['emotion'].replace('8','0')  #(label number 0 represents surprise emotion(08))

# EXTRACTING DEPENDENT AND INDEPENDENT VARIABLE
x_train1 = df5.iloc[:, 1:-1].values
y_train1 = df5.iloc[:, 60].values
x_test1 = df6.iloc[:, 1:-1].values
y_test1 = df6.iloc[:, 60].values
```

```python
# SVM
# Making the SVM Classifer
Classifier = SVC(kernel="rbf")

# Training the model on the training data and labels
Classifier.fit(x_train1, y_train1)

# Using the model to predict the labels of the test data
y_pred = Classifier.predict(x_test1)

# Evaluating the accuracy of the model using the sklearn functions
accuracy2 = (accuracy_score(y_test1,y_pred))*100

# Printing the results
print("Accuracy for SVM is:",accuracy2)
```

```python
# Using XGBoost
# ------------------------------------------------------------->>>

# Applying XGBoost
xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(x_train1,y_train1)
preds = xgb_classifier.predict(x_test1)
xgb_score2=(accuracy_score(y_test1,preds))*100
print('Accuracy Score : ',xgb_score2)
```

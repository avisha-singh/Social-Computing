```python
import re
import pprint
import pandas as pd
import glob
import os

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
path = "/content/gdrive/MyDrive/Data Vr + Meditation/csv"
csv_files = glob.glob(path + "/*.csv")
Avg1=[]
Avg2=[]
Avg3=[]
Acc1=[]
Acc2=[]
Acc3=[]
name=[]
falr1=[]
missper1=[]
matchper1=[]
falr2=[]
missper2=[]
matchper2=[]
falr3=[]
missper3=[]
matchper3=[]
nummissone=[]
nummisstwo=[]
nummissthree=[]
numcorrmatchone=[]
numcorrmatchtwo=[]
numcorrmatchthree=[]
numfalsealrmone=[]
numfalsealrmtwo=[]
numfalsealrmthree=[]
numtotmatchone=[]
numtotmatchtwo=[]
numtotmatchthree=[]
numnomatchone=[]
numnomatchtwo=[]
numnomatchthree=[]

for i in csv_files:
    file_name = os.path.basename(i)
    pth=(os.path.splitext(file_name)[0])
    name.append(pth[6:])
    df=pd.read_csv(i)

    one_back=df[df['block'] =='t1']

    false_alarm_one=one_back[(one_back.resp1 == '<lshift>') & (one_back.corr1 == 0)]
    numfalsealrmone.append(false_alarm_one.shape[0])

    match_one=one_back[(one_back.resp1=='<lshift>')&(one_back.corr1==1)]
    numcorrmatchone.append(match_one.shape[0])

    miss_one=one_back[(one_back.resp1=='<timeout>')&(one_back.corr1==0)]
    nummissone.append(miss_one.shape[0])

    cateogary_one=one_back[(one_back.resp1=='<timeout>')&(one_back.corr1==1)]

    total_match_oneback=one_back['d1'].value_counts()[1]
    numtotmatchone.append(total_match_oneback)

    nomatch_oneback=41-total_match_oneback
    numnomatchone.append(nomatch_oneback)

    false_alarm_perc_one= ((false_alarm_one.shape[0])/nomatch_oneback)*100
    falr1.append(false_alarm_perc_one)

    miss_perc_one=((miss_one.shape[0])/total_match_oneback)*100
    missper1.append(miss_perc_one)

    match_perc_one=((match_one.shape[0])/total_match_oneback)*100
    matchper1.append(match_perc_one)

    Accuracy_one= (((match_one.shape[0])+(cateogary_one.shape[0]))/one_back.shape[0])*100
    Acc1.append(Accuracy_one)

    Avg_time_one=one_back['rta'].mean()
    Avg1.append(Avg_time_one)



    two_back=df[df['block'] =='t2']

    false_alarm_two=two_back[(two_back.resp1 == '<lshift>') & (two_back.corr1 == 0)]
    numfalsealrmtwo.append(false_alarm_two.shape[0])

    match_two=two_back[(two_back.resp1=='<lshift>')&(two_back.corr1==1)]
    numcorrmatchtwo.append(match_two.shape[0])

    miss_two=two_back[(two_back.resp1=='<timeout>')&(two_back.corr1==0)]
    nummisstwo.append(miss_two.shape[0])

    cateogary_two=two_back[(two_back.resp1=='<timeout>')&(two_back.corr1==1)]

    total_match_twoback=one_back['d1'].value_counts()[1]
    numtotmatchtwo.append(total_match_twoback)

    nomatch_twoback=42-total_match_twoback
    numnomatchtwo.append(nomatch_twoback)

    false_alarm_perc_two= ((false_alarm_two.shape[0])/nomatch_twoback)*100
    falr2.append(false_alarm_perc_two)

    miss_perc_two=((miss_two.shape[0])/total_match_twoback)*100
    missper2.append(miss_perc_two)

    match_perc_two=((match_two.shape[0])/total_match_twoback)*100
    matchper2.append(match_perc_two)

    Accuracy_two= (((match_two.shape[0])+(cateogary_two.shape[0]))/two_back.shape[0])*100
    Acc2.append(Accuracy_two)

    Avg_time_two=two_back['rta'].mean()
    Avg2.append(Avg_time_two)



    three_back=df[df['block'] =='t3']

    false_alarm_three=three_back[(three_back.resp1 == '<lshift>') & (three_back.corr1 == 0)]
    numfalsealrmthree.append(false_alarm_three.shape[0])

    match_three=three_back[(three_back.resp1=='<lshift>')&(three_back.corr1==1)]
    numcorrmatchthree.append(match_three.shape[0])

    miss_three=three_back[(three_back.resp1=='<timeout>')&(three_back.corr1==0)]
    nummissthree.append(miss_three.shape[0])

    cateogary_three=three_back[(three_back.resp1=='<timeout>')&(three_back.corr1==1)]

    total_match_threeback=three_back['d1'].value_counts()[1]
    numtotmatchthree.append(total_match_threeback)

    nomatch_threeback=43-total_match_threeback
    numnomatchthree.append(nomatch_threeback)


    false_alarm_perc_three= ((false_alarm_three.shape[0])/nomatch_threeback)*100
    falr3.append(false_alarm_perc_three)

    miss_perc_three=((miss_three.shape[0])/total_match_threeback)*100
    missper3.append(miss_perc_three)

    match_perc_three=((match_three.shape[0])/total_match_threeback)*100
    matchper3.append(match_perc_three)

    Accuracy_three= (((match_three.shape[0])+(cateogary_three.shape[0]))/three_back.shape[0])*100
    Acc3.append(Accuracy_three)

    Avg_time_three=three_back['rta'].mean()
    Avg3.append(Avg_time_three)

csv1=pd.DataFrame(Avg1)
csv2=pd.DataFrame(Acc1)
csv3=pd.DataFrame(falr1)
csv4=pd.DataFrame(missper1)
csv5=pd.DataFrame(matchper1)
csv6=pd.DataFrame(nummissone)
csv7=pd.DataFrame(numcorrmatchone)
csv8=pd.DataFrame(numfalsealrmone)
csv9=pd.DataFrame(numtotmatchone)
csv10=pd.DataFrame(numnomatchone)
csv11=pd.DataFrame(Avg2)
csv12=pd.DataFrame(Acc2)
csv13=pd.DataFrame(falr2)
csv14=pd.DataFrame(missper2)
csv15=pd.DataFrame(matchper2)
csv16=pd.DataFrame(nummisstwo)
csv17=pd.DataFrame(numcorrmatchtwo)
csv18=pd.DataFrame(numfalsealrmtwo)
csv19=pd.DataFrame(numtotmatchtwo)
csv20=pd.DataFrame(numnomatchtwo)
csv21=pd.DataFrame(Avg3)
csv22=pd.DataFrame(Acc3)
csv23=pd.DataFrame(falr3)
csv24=pd.DataFrame(missper3)
csv25=pd.DataFrame(matchper3)
csv26=pd.DataFrame(nummissthree)
csv27=pd.DataFrame(numcorrmatchthree)
csv28=pd.DataFrame(numfalsealrmthree)
csv29=pd.DataFrame(numtotmatchthree)
csv30=pd.DataFrame(numnomatchthree)


comb= pd.concat([csv1,csv2,csv3,csv4,csv5,csv6,csv7,csv8,csv9,csv10,csv11,csv12,csv13,csv14,csv15,csv16,csv17,csv18,csv19,csv20,csv21,csv22,csv23,csv24,csv25,csv26,csv27,csv28,csv29,csv30], axis=1)
comb.columns =['Avg_time_one', 'Accuracy_one','False Alarm Percentage_one','Miss Percentage_one','Match Percentage_one','No.of missed items_one','No. of correct match_one','No. of false alarm_one','Total match_one','Nomatch_one','Avg_time_two','Accuracy_two','False Alarm Percentage_two','Miss Percentage_two','Match Percentage_two','No.of missed items_two','No. of correct match_two','No. of false alarm_two','Total match_two','Nomatch_two','Avg_time_three','Accuracy_three','False Alarm Percentage_three','Miss Percentage_three','Match Percentage_three','No.of missed items_three','No. of correct match_three','No. of false alarm_three','Total match_three','Nomatch_three']
comb.index=name
comb.to_csv("N_back_result.csv")
```

```python
path = "/content/gdrive/MyDrive/Data Vr + Meditation/text"
txt_files = glob.glob(path + "/*.txt")
c_avgtime=[]
i_avgtime=[]
n_avgtime=[]
c_acc=[]
i_acc=[]
n_acc=[]
name1=[]
for i in txt_files:
    file_name = os.path.basename(i)
    pth1=(os.path.splitext(file_name)[0])
    name1.append(pth1[27:])
    f = open(i, "r")
    cont = f.read()
    list=re.split('; |, |\*|\n',cont)
    c_meantime=(list[24][8:16])
    i_meantime=(list[25][8:16])
    n_meantime=(list[26][8:16])
    c_accuracy=(list[30][8:16])
    i_accuracy=(list[31][8:18])
    n_accuracy=(list[32][8:18])

    c_avgtime.append(c_meantime)
    i_avgtime.append(i_meantime)
    n_avgtime.append(n_meantime)
    c_acc.append(c_accuracy)
    i_acc.append(i_accuracy)
    n_acc.append(n_accuracy)

DF1=pd.DataFrame(c_avgtime)
DF2=pd.DataFrame(c_acc)
DF3=pd.DataFrame(i_avgtime)
DF4=pd.DataFrame(i_acc)
DF5=pd.DataFrame(n_avgtime)
DF6=pd.DataFrame(n_acc)
COMB= pd.concat([DF1,DF2,DF3,DF4,DF5,DF6], axis=1)
COMB.columns =['Avg_time_C', 'Accuracy_C','Avg_time_I','Accuracy_I','Avg_time_N','Accuracy_N']
COMB.index=name1
COMB.to_csv("Stroop_result.csv")
```

```python
from pandas.core.describe import DataFrameDescriber
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv('/content/gdrive/MyDrive/Data Vr + Meditation/Final EEG.csv')
DF= pd.read_csv('/content/gdrive/MyDrive/Data Vr + Meditation/N_back_result.csv')
data=pd.read_csv('/content/gdrive/MyDrive/Data Vr + Meditation/Stroop_result.csv')
```

```python
# EEG for 1 and 2
df3 = df[df['Unnamed: 0'].str.contains('_1')]
df4 = df[df['Unnamed: 0'].str.contains('_2')]
df1=df3.drop([7,12,35,61])
df1 = df1.sample(n=30)
df2=df4.drop([3,14,59])
df2 = df2.sample(n=30)

df1.to_csv('/content/gdrive/MyDrive/Data Vr + Meditation/EEG BEFORE.csv')
df2.to_csv('/content/gdrive/MyDrive/Data Vr + Meditation/EEG AFTER.csv')

# Frontal Alpha
df1['F_Alpha'] = df1['AF7_Alpha'] + df1['AF8-Alpha']
df2['F_Alpha'] = df2['AF7_Alpha'] + df2['AF8-Alpha']
data_group1 = df1['F_Alpha'].values
data_group2 = df2['F_Alpha'].values
ttest=stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
print('Frontal Alpha T-Test: ',ttest)
print('Frontal Alpha-1: ',data_group1.mean())
print('Frontal Alpha-2: ',data_group2.mean())
print()

# Perital Alpha
df1['P_Alpha'] = df1['TP9_Alpha'] + df1['TP10_Alpha']
df2['P_Alpha'] = df2['TP9_Alpha'] + df2['TP10_Alpha']
data_group3 = df1['P_Alpha'].values
data_group4 = df2['P_Alpha'].values
ttest1=stats.ttest_ind(a=data_group3, b=data_group4, equal_var=True)
print('Perital Alpha T-Test: ',ttest1)
print('Perital Alpha-1: ',data_group3.mean())
print('Perital Alpha-2: ',data_group4.mean())
print()

# Temporal Theta
df1['T_Theta'] = df1['TP9_Theta'] + df1['TP10_Theta']
df2['T_Theta'] = df2['TP9_Theta'] + df2['TP10_Theta']
data_group5 = df1['T_Theta'].values
data_group6 = df2['T_Theta'].values
ttest2=stats.ttest_ind(a=data_group5, b=data_group6, equal_var=True)
print('Temporal Theta T-Test: ',ttest2)
print('Temporal Theta-1: ',data_group5.mean())
print('Temporal Theta-2: ',data_group6.mean())
print()

# PLOTTING GRAPH
barWidth = 0.15
fig = plt.subplots(figsize =(7,4))

one_1=[data_group1.mean(),data_group3.mean(),data_group5.mean()]
three_2=[data_group2.mean(),data_group4.mean(),data_group6.mean()]

br1 = np.arange(len(one_1))
br2 = [x + barWidth for x in br1]

plt.bar(br1,one_1,color='pink',width=barWidth,label='1')
plt.bar(br2,three_2,color='b',width=barWidth,label='2')
plt.xlabel('\nEEG',fontsize=15)
plt.ylabel('In (microVolt)^2',fontsize=15)
plt.xticks([r + barWidth for r in range(len(one_1))],['Frontal Alpha\np=0.46','Perital Alpha\np=0.37','Temporal Theta\np=0.78'])
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
# Stroop for 1 and 2

data1=data.drop(['Time'],axis=1)
df1 = data1[data1['Unnamed: 0'].str.contains('_1')]
df3=df1.drop([1,20,25,45,59,64])
df3=df3.sample(n=30)
df2 = data1[data1['Unnamed: 0'].str.contains('_2')]
df4=df2.drop([6,14,17,24,42,61,62,66])
df4=df4.sample(n=30)

df3.to_csv('/content/gdrive/MyDrive/Data Vr + Meditation/STROOP BEFORE.csv')
df4.to_csv('/content/gdrive/MyDrive/Data Vr + Meditation/STROOP AFTER.csv')

# Average time C
data_group1 = df3.iloc[:,1].values
data_group2 = df4.iloc[:,1].values
ttest=stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
print('Average time C T-Test: ',ttest)
print('Average time C-1: ',data_group1.mean())
print('Average time C-2: ',data_group2.mean())
print()

# Average time I
data_group3 = df3.iloc[:,3].values
data_group4 = df4.iloc[:,3].values
ttest1=stats.ttest_ind(a=data_group3, b=data_group4, equal_var=True)
print('Average time I T-Test: ',ttest1)
print('Average time I-1: ',data_group3.mean())
print('Average time I-2: ',data_group4.mean())
print()

# Average time N
data_group5 = df3.iloc[:,5].values
data_group6 = df4.iloc[:,5].values
ttest2=stats.ttest_ind(a=data_group5, b=data_group6, equal_var=True)
print('Average time N T-Test: ',ttest2)
print('Average time N-1: ',data_group5.mean())
print('Average time N-2: ',data_group6.mean())
print()

# Accuracy C
data_group7 = df3.iloc[:,2].values
data_group8 = df4.iloc[:,2].values
ttest3=stats.ttest_ind(a=data_group7, b=data_group8, equal_var=True)
print('Accuracy C T-Test: ',ttest3)
print('Accuracy C-1: ',data_group7.mean())
print('Accuracy C-2: ',data_group8.mean())
print()

# Accuracy I
data_group9 = df3.iloc[:,4].values
data_group10 = df4.iloc[:,4].values
ttest4=stats.ttest_ind(a=data_group9, b=data_group10, equal_var=True)
print('Accuracy I T-Test: ',ttest4)
print('Accuracy I-1: ',data_group9.mean())
print('Accuracy I-2: ',data_group10.mean())
print()

# Accuracy N
data_group11 = df3.iloc[:,6].values
data_group12 = df4.iloc[:,6].values
ttest5=stats.ttest_ind(a=data_group11, b=data_group12, equal_var=True)
print('Accuracy N T-Test: ',ttest5)
print('Accuracy N-1: ',data_group11.mean())
print('Accuracy N-2: ',data_group12.mean())
print()

# PLOTTING GRAPH (AVERAGE TIME)
barWidth = 0.15
fig = plt.subplots(figsize =(7,4))

one_1=[data_group1.mean(),data_group3.mean(),data_group5.mean()]
three_2=[data_group2.mean(),data_group4.mean(),data_group6.mean()]

br1 = np.arange(len(one_1))
br2 = [x + barWidth for x in br1]

plt.bar(br1,one_1,color='pink',width=barWidth,label='1')
plt.bar(br2,three_2,color='b',width=barWidth,label='2')
plt.title('\nStroop',fontsize=18)
plt.xlabel('\nAverage Time',fontsize=15)
plt.ylabel('In seconds',fontsize=15)
plt.xticks([r + barWidth for r in range(len(one_1))],['C\np=0.67','I\np=0.65','N\np=0.59'])
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()

# PLOTTING GRAPH (ACCURACY)
barWidth = 0.15
fig = plt.subplots(figsize =(7,4))

one=[data_group7.mean(),data_group9.mean(),data_group11.mean()]
three=[data_group8.mean(),data_group10.mean(),data_group12.mean()]

br3 = np.arange(len(one))
br4 = [x + barWidth for x in br3]


plt.bar(br3,one,color='pink',width=barWidth,label='1')
plt.bar(br4,three,color='b',width=barWidth,label='2')
plt.title('\nStroop',fontsize=18)
plt.xlabel('\nAccuracy',fontsize=15)
plt.ylabel('Values',fontsize=15)
plt.xticks([r + barWidth for r in range(len(one))],['C\np=0.48','I\np=0.46','N\np=0.99'])
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
